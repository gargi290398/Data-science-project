```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('data1.csv')
df.head()
```

```python
print(df.columns)
```

```python

plt.scatter(df['Year'], df['Cereals and products'])
plt.xlabel('Year')
plt.ylabel('Cereals and products')
plt.title('Scatter plot of Year vs Cereals and products')
plt.show()

plt.scatter(df['Year'], df['Meat and fish'])
plt.xlabel('Year')
plt.ylabel('Meat and fish')
plt.title('Scatter plot of Year vs Meat and fish')
plt.show()

plt.scatter(df['Year'], df['Vegetables'])
plt.xlabel('Year')
plt.ylabel('Vegetables')
plt.title('Scatter plot of Year vs Vegetables')
plt.show()

plt.scatter(df['Year'], df['Education'])
plt.xlabel('Year')
plt.ylabel('Education')
plt.title('Scatter plot of Year vs Education')
plt.show()
```

```python
df_cleaned = df.dropna(subset=['Cereals and products', 'Meat and fish', 'Vegetables', 'Health','Education'])
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(15, 10))

axes[0, 0].boxplot(df_cleaned['Cereals and products'], showfliers=False)
axes[0, 0].set_title('Cereals and products')

axes[0, 1].boxplot(df_cleaned['Meat and fish'], showfliers=False)
axes[0, 1].set_title('Meat and fish')

axes[1, 0].boxplot(df_cleaned['Vegetables'], showfliers=False)
axes[1, 0].set_title('Vegetables')

axes[1, 1].boxplot(df_cleaned['Education'], showfliers=False)
axes[1, 1].set_title('Education')
plt.tight_layout()
plt.show()
```

```python
yearly_avg = df.groupby('Year')['Cereals and products'].mean()

yearly_avg.plot(kind='line')
plt.xlabel('Year')
plt.ylabel('Cereals and products (Average)')
plt.title('Line plot of Year vs Cereals and products (Average)')
plt.show()
```

```python
yearly_avg = df.groupby('Year')['Meat and fish'].mean()

yearly_avg.plot(kind='line')
plt.xlabel('Year')
plt.ylabel('Meat and fish (Average)')
plt.title('Line plot of Year vs Meat and fish (Average)')
plt.show()

yearly_avg = df.groupby('Year')['Vegetables'].mean()

yearly_avg.plot(kind='line')
plt.xlabel('Year')
plt.ylabel('Vegetables (Average)')
plt.title('Line plot of Year vs Vegetables (Average)')
plt.show()


yearly_avg = df.groupby('Year')['Education'].mean()

yearly_avg.plot(kind='line')
plt.xlabel('Year')
plt.ylabel('Education (Average)')
plt.title('Line plot of Year vs Education (Average)')
plt.show()
```
